```python
from sklearn import datasets

dados.columns = dados.columns.str.strip()
```

```python
dados.isnull().value_counts()
```

```python
dados_x = dados.drop(columns=["ID", "default.payment.next.month"])
dados_y = dados["default.payment.next.month"]

```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(dados_x)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, dados_y, test_size = 0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import joblib

os.makedirs('models', exist_ok=True)


joblib.dump(model, "models/rf_model.pkl")
joblib.dump(model, "models/scaler.pkl")
```
